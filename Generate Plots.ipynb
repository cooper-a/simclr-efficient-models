```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "plots/"

def plot_results(results, filename, path, title):
    plt.plot(results['test_acc@1'], label='Top-1 Test Accuracy')
    
    plt.title(title + ' Top-1 Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Top-1 Accuracy')
    plt.legend()
    plt.savefig(path + 'top1_accuracy_' + filename)
    plt.show()

    plt.plot(results['test_acc@5'], label='Top-5 Test Accuracy')
    plt.title(title + ' Top-5 Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Top-5 Accuracy')
    plt.legend()
    plt.savefig(path + 'top5_accuracy_' + filename)
    plt.show()

    plt.plot(results['train_loss'], label='train loss')
    plt.title(title + ' Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig(path + 'loss_' + filename)
    plt.show()
```

```python
df = pd.read_csv('final_results/mobilenetv3small_1024_batch_1000_epoch/128_0.5_200_1024_1000_statistics.csv')
plot_results(df.iloc[::20, :], "mobilenetv3small_1024_batch_1000_epoch", path, "MobileNetV3 Small Batch Size 1024")
```

```python
df = pd.read_csv('final_results/mobilenetv3large_256_batch_1000_epoch/128_0.5_200_256_1000_statistics.csv')
plot_results(df.iloc[::20, :], "mobilenetv3large_256_batch_1000_epoch", path, "MobileNetV3 Small")
```

```python
df = pd.read_csv('final_results/mobilenetv3large_1024_batch_1000_epoch/128_0.5_200_1024_1000_statistics.csv')
plot_results(df.iloc[::20, :], "mobilenetv3large_1024_batch_1000_epoch", path, "MobileNetV3 Large")
```

```python
df = pd.read_csv('final_results/resnet18_512_batch_1000_epoch/128_0.5_200_512_1000_statistics.csv')
plot_results(df.iloc[::20, :], "resnet18_512_batch_1000_epoch", path, "ResNet18")
```

```python
# Plot batch size comparison
df_256 = pd.read_csv('final_results/mobilenetv3large_256_batch_1000_epoch/128_0.5_200_256_1000_statistics.csv')
df_512 = pd.read_csv('final_results/mobilenetv3large_512_batch_1000_epoch/128_0.5_200_512_1000_statistics.csv')
df_1024 = pd.read_csv('final_results/mobilenetv3large_1024_batch_1000_epoch/128_0.5_200_1024_1000_statistics.csv')

df_256 = df_256.iloc[::20, :]
df_512 = df_512.iloc[::20, :]
df_1024 = df_1024.iloc[::20, :]

plt.plot(df_256['test_acc@1'], label='256 batch size')
plt.plot(df_512['test_acc@1'], label='512 batch size')
plt.plot(df_1024['test_acc@1'], label='1024 batch size')
    
plt.title('Batch Size Comparison of MobileNetV3 Large')
plt.xlabel('Epoch')
plt.ylabel('Top-1 Accuracy')
plt.legend()
plt.savefig(path + 'top1_accuracy_batch_comparison')
plt.show()
```

```python
# Training curves

df_1024_large = pd.read_csv('final_results/mobilenetv3large_1024_batch_1000_epoch/128_0.5_200_1024_1000_statistics.csv')
df_1024_small = pd.read_csv('final_results/mobilenetv3small_1024_batch_1000_epoch/128_0.5_200_1024_1000_statistics.csv')
df_1024_large = df_1024_large.iloc[::20, :]
df_1024_small = df_1024_small.iloc[::20, :]

plt.plot(df_1024_large['train_loss'], label='MobilenetV3 Large')
plt.plot(df_1024_small['train_loss'], label='MobilenetV3 Small')

plt.title('Training Loss Comparison of MobileNetV3 Small vs Large (Batch size = 1024)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.savefig(path + 'loss_compare')
plt.show()
```
